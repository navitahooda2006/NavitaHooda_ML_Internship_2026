QUES 8: Scatter Plot Analysis
Create scatter plots of:
1. rainfall_mm vs yield_ton_per_hectare
2. fertilizer_kg vs yield_ton_per_hectare

Based on the plots:

 Which feature appears to have a stronger relationship with yield?

```python
#ques 8:
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(data=df, x='rainfall_mm', y='yield_ton_per_hectare', ax=axes[0], color='skyblue', alpha=0.6)
axes[0].set_title('Rainfall vs. Crop Yield')
axes[0].set_xlabel('Rainfall (mm)')
axes[0].set_ylabel('Yield (ton per hectare)')

sns.scatterplot(data=df, x='fertilizer_kg', y='yield_ton_per_hectare', ax=axes[1], color='mediumpurple', alpha=0.6)
axes[1].set_title('Fertilizer vs. Crop Yield')
axes[1].set_xlabel('Fertilizer (kg)')
axes[1].set_ylabel('Yield (ton per hectare)')

plt.tight_layout()
plt.show()

corr_rainfall = df['rainfall_mm'].corr(df['yield_ton_per_hectare'])
corr_fertilizer = df['fertilizer_kg'].corr(df['yield_ton_per_hectare'])

print(f"Correlation between Rainfall and Yield: {corr_rainfall:.4f}")
print(f"Correlation between Fertilizer and Yield: {corr_fertilizer:.4f}")

# TEXT ANALYTICAL ANSWER (Observations)
# Based on the plots and correlation coefficients:
# -> Fertilizer ('fertilizer_kg') has a significantly stronger relationship with yield.
#
# Reason:
# 1. The 'Fertilizer vs. Crop Yield' scatter plot displays a clear, visible upward
#    linear trend (as fertilizer increases, yield consistently increases).
# 2. In contrast, the 'Rainfall vs. Crop Yield' plot shows a highly dispersed,
#    cloud-like pattern with no distinct upward or downward slope, indicating
#    a very weak or negligible linear relationship.
```

QUES 9: Correlation Analysis

 Generate a correlation matrix for numerical features.

 Create a heatmap.

 Identify the top three features most correlated with crop yield.

```python
#ques 9:
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df.corr(numeric_only=True)
print("Correlation Matrix: ")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".4f", square=True, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

yield_correlation = correlation_matrix['yield_ton_per_hectare'].drop('yield_ton_per_hectare')

top_3_features = yield_correlation.abs().nlargest(3).index

print("Top three features most correlated with crop yield:")
for rank, feature in enumerate(top_3_features, 1):
    actual_corr_val = yield_correlation[feature]
    print(f" {rank}. '{feature}' (Correlation coefficient = {actual_corr_val:.4f})")
```

QUES 10: Group-Based Analysis

Calculate the average yield for:

 Each crop type

 Each soil type

Which crop and soil type have the highest average yield?

```python
#ques 10:
crop_yield_means = df.groupby('crop_type')['yield_ton_per_hectare'].mean()
print("Average Yield by Crop Type")
print(crop_yield_means)

soil_yield_means = df.groupby('soil_type')['yield_ton_per_hectare'].mean()
print("Average Yield by Soil Type")
print(soil_yield_means)

highest_crop = crop_yield_means.idxmax()
highest_crop_val = crop_yield_means.max()

highest_soil = soil_yield_means.idxmax()
highest_soil_val = soil_yield_means.max()

print(f"The crop type with the highest average yield is: '{highest_crop}' ({highest_crop_val:.4f} tons/hectare)")
print(f"The soil type with the highest average yield is: '{highest_soil}' ({highest_soil_val:.4f} tons/hectare)")
```

QUES 11: Feature Encoding

The dataset contains categorical variables.

 Identify the categorical columns.

 Convert them into numerical form using One-Hot Encoding.

 Display the first five rows of the transformed dataset.

```python
#ques 11:
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical Columns Identified")
print(categorical_cols)

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=int)

print("First 5 rows of the transformed dataset:")
df_encoded.head()
```

QUES 12: Feature Selection

Separate:

 Input features (X)

 Target variable (y)

Specify which column is being used as the target variable.

```python
#ques 12:
y = df_encoded['yield_ton_per_hectare']

X = df_encoded.drop(columns=['yield_ton_per_hectare'])

print("Feature Selection Details")
print(f"Target variable (y): 'yield_ton_per_hectare'")
print(f"Input features matrix (X) shape: {X.shape} (Features: {list(X.columns)})")
```

QUES 13: Train-Test Split

Split the dataset into:

 80% Training Data

 20% Testing Data

Display the shape of:

 X_train

 X_test

 y_train

 y_test

```python
#ques 13:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print("Train-Test Split Dimensions")
print(f"• Shape of X_train : {X_train.shape} (1200 rows, 11 features)")
print(f"• Shape of X_test  : {X_test.shape}  (300 rows, 11 features)")
print(f"• Shape of y_train : {y_train.shape} (1200 target labels)")
print(f"• Shape of y_test  : {y_test.shape}  (300 target labels)")
```

QUES 14: Linear Regression Model

 Train a Linear Regression model.

 Display the model coefficients and intercept.

 Which feature has the highest positive coefficient?

```python
#ques 14:
from sklearn.linear_model import LinearRegression
import pandas as pd

model = LinearRegression()
model.fit(X_train, y_train)

print("Linear Regression Model Summary")
print(f"Intercept: {model.intercept_:.4f}\n")

coefficients = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': model.coef_
})
print("Model Coefficients:")
print(coefficients.to_string(index=False))

highest_positive_idx = coefficients['Coefficient'].idxmax()
best_feature = coefficients.loc[highest_positive_idx, 'Feature']
best_feature_coef = coefficients.loc[highest_positive_idx, 'Coefficient']

print(f"• The feature with the highest positive coefficient is: '{best_feature}'")
print(f"  Coefficient value: {best_feature_coef:.4f}")
```
